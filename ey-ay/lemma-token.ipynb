```python
import spacy
text = open('../input/genesis/Genesis.txt', 'r')

text= text.read()

text = text.replace(",", "")
text = text.replace(".", "")
text = text.replace("\n", "")
text = text.replace(";", "")
text = text.replace(":", "")
    

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)




token_list = [token for token in doc if not token.is_stop]
# token_list

lemmas = [
    f"Token: {token}, lemma: {token.lemma_}"
    for token in token_list
]
lemmas


```

```python
def listToString(input): 
    str1 = " " 
    return (str1.join(input))
```

```python
valString = listToString(lemmas)

freq_num = valString.count("God")
print(freq_num/2) # each word is signified twice, one as a token, one as a lemma. 
```
